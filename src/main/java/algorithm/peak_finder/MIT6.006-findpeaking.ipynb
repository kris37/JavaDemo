## find peaking
* 给出一个数组，查找凸点即 array[i]>array[i - 1] && array[i] > array[i + 1]

```python
array1D = [1,3,5,7,3,2,4,3,2,5,2,6,7]
# pic
import matplotlib.pyplot as plt
import numpy as np
my_x_ticks = np.arange(0, len(array1D) + 1, 1)
plt.xticks(my_x_ticks)
plt.plot(range(0,len(array1D)), array1D, marker='*', ms=10)
plt.show()
```

```python
# 采用2分法进行查找一个peak 点
def find1Dpeak(arr):
    left = 0
    right = len(arr) - 1
    cur = 0
    while(left < right):
        cur = (left + right) >> 1
        if arr[cur] < arr[cur + 1]:
            left = cur + 1
        else:
            right = cur
    return ('value:'+str(arr[left]),'index:'+str(left))
```

```python
find1Dpeak(array1D)
```
